```python
#Import Libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.tools.sm_exceptions as sm_except
from statsmodels.regression.rolling import RollingOLS
```

```python
#Variables 
filename = '../data/InterviewTestData.xlsx'
outlier_detect_value = 4 #In standard deviations
benchmark = 'MSCI World'
```

```python
#Load Files
port_returns = pd.read_excel(filename, skiprows=1, sheet_name='input_portfolio').dropna(axis=1, how='all')
factor_returns = pd.read_excel(filename, skiprows=1, sheet_name='input_factors').dropna(axis=1, how='all')
```

```python
#Set Indices
port_returns['Date'] = pd.to_datetime(port_returns['Date'])
port_returns = port_returns.set_index(keys='Date')
factor_returns['Date'] = pd.to_datetime(factor_returns['Date'])
factor_returns = factor_returns.set_index(keys='Date')
```

```python
#Fix Erroneous Data
port_returns = port_returns.replace('---', np.NaN)
port_returns = port_returns.astype('float')
```

```python
# Ffill as per instructions, perhaps 0 might be better
factor_returns = factor_returns.ffill()
port_returns = port_returns.ffill()

benchmark = 'MSCI World'
factors = factor_returns.columns.drop(benchmark)
portfolios = port_returns.columns.drop(benchmark)
```

## Item 1: Outlier Detection
Visualising the data with boxplots we can see outliers in P2, P1 and MSCI Momentum. However, we will use z-scores to systematically determine outliers

```python
port_returns.plot(kind='box')
factor_returns.plot(kind='box')
```

```python
#Calculate Z-Score [(x - mu) / sigma]
factor_zscore = (factor_returns - factor_returns.mean()) / factor_returns.std()
port_zscore = (port_returns - port_returns.mean()) / port_returns.std()
```

```python
ax1 = sns.heatmap(factor_zscore)

y_dates = factor_zscore.index.strftime('%Y-%m-%d')
ax1.set_yticklabels(labels=y_dates)

plt.show()

ax2 = sns.heatmap(port_zscore)
y_dates = port_zscore.index.strftime('%Y-%m-%d')
ax2.set_yticklabels(labels=y_dates)

plt.show()
```

If one assumes a normal distribution then a 4 sigma move is extremely unlikely,
3sigma covers ~ 99.99%. Given the data this is reasonable for now. We would need to check
the veracity of these moves. We set these returns to 0, assuming bad data. 
Winsorisation would be another possibility, and replace these outliers with average etc.

```python

# If one assumes a normal distribution then a 4 sigma move is extremely unlikely,
# #3sigma covers ~ 99.99%. Given the data this is reasonable for now. We would need to check
# the veracity of these moves. We set these returns to 0, assuming bad data. Winsorisation would
# be another possibility, and replace these outliers with average etc.
outlier_mask = np.abs(factor_zscore) > outlier_detect_value
factor_outliers = factor_returns[outlier_mask].dropna(axis=0, how='all').dropna(axis=1, how='all')
factor_returns.iloc[outlier_mask] = 0


outlier_mask = np.abs(port_zscore) > outlier_detect_value
port_outliers = port_returns[outlier_mask].dropna(axis=0, how='all').dropna(axis=1, how='all')
port_returns.iloc[outlier_mask] = 0
```

## Item 6: OLS
We initially do a regression over the full period to make a point on the significance of some factors to the portfolio. Some points to note. For <b>P1</b>, Value and Momentum are not significant and Value is of a similar scale to Quality which is significant. Removing these may improve the fit (using AIC) of the model. For <b>P2</b> we note that Value, Momentum and Quality are not significant which is similar for <b>P5</b>. For <b>P3</b> Momentum and Quality are not significant and <b>P4</b> Momentum is not significant. We will continue with the analysis as planned but suggest that the combination of factors for a given portfolio may need to be reviewed, particularly Momentum across the board. Furthermore, we have not specifically tested for heteroskedasticity but on visualising the residuals we note that this may be an issue that requires further investigation. The tables below are an unfortunately verbose summary of the text. 

```python
X = factor_returns.loc[port_returns.index, factors]
X = sm.add_constant(X)

for portfolio in portfolios:
    y = port_returns[portfolio]
    model = sm.OLS(y, X)
    y_hat = model.fit()
    print(y_hat.summary())
```

### Item 6a: We determine the portfolio/manager alpha/idiosyncratic return as the residual as the model. So if the model is r = XB + e we take e as our idiosyncratic return (const included), r-XB = e

```python
ax_dict = {}
for factor in factors:
    f, ax = plt.subplots(figsize=(11, 9))
    ax_dict[factor] = ax

f, ax = plt.subplots(figsize=(11, 9))
ax_dict['Alpha'] = ax

for portfolio in portfolios:
    y = port_returns[portfolio]
    model = RollingOLS(y, X, window=36, min_nobs=36)
    rolling_res = model.fit()
    params = rolling_res.params
    # We don't use these explicitly but note they may be useful given the commentary above over the whole sample
    rolling_res.pvalues
    rolling_res.ssr.rename(portfolio).plot(title='Alpha', ax=ax_dict['Alpha'], legend=True)
        
    for factor in factors:
        params[factor].rename(portfolio).plot(ax=ax_dict[factor], title=factor, legend=True)
    
```

## Item 8: Forecast portfolio returns
We could use some timeseries modelling ARIMA/ARDL etc but for the purposes of this exercise we will simply simulate future returns given the distribution of past returns. We assume a normal ditribution. 

```python
mu = port_returns[portfolios].mean()
print(mu)
sigma = port_returns[portfolios].std()
print(sigma)
```

```python
periods = 12
n = 100
for portfolio in portfolios:
    r = np.random.normal(loc=mu[portfolio], scale=sigma[portfolio], size=n)
    df = 1 + pd.DataFrame(np.reshape(r, newshape=(periods, n)))
    (df.prod() - 1).mean()
    (df.prod() - 1).std()
    
```
